## Langkah 1: Import Library

```python
# Import library yang dibutuhkan
import cv2
import matplotlib.pyplot as plt
```


* `cv2`: Digunakan untuk membaca dan memproses citra.
* `numpy`: Membantu dengan perhitungan numerik (jika diperlukan).
* `matplotlib.pyplot`: Untuk memvisualisasikan histogram.

## Langkah 2: Membaca Citra Grayscale

```python
# Membaca citra grayscale
image_path = '/content/drives/MyDrive/download (4).jpeg'  # Ganti dengan path gambar kamu
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Periksa apakah gambar berhasil dibaca
if image is None:
    raise ValueError("Gambar tidak dapat dibaca. Pastikan path gambar benar.")

# Menampilkan citra grayscale
plt.imshow(image, cmap='gray')
plt.title("Citra Grayscale")
plt.axis("off")
plt.show()
```

* `cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)` membaca citra dalam mode grayscale.
* Jika path salah, akan muncul error yang meminta perbaikan.
* `plt.imshow()` menampilkan citra dalam colormap grayscale.

## Langkah 3: Menghitung Histogram

```python
# Menghitung histogram citra
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Menampilkan hasil histogram (tanpa normalisasi)
plt.plot(hist)
plt.title("Histogram Citra Grayscale")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")
plt.xlim([0, 256])
plt.grid()
plt.show()
```

Fungsi `cv2.calcHist()` menghitung histogram dengan parameter:
* `[image]`: Citra yang dihitung histogramnya.
* `0`: Channel (0 = grayscale).
* `None`: Mask (tidak menggunakan mask).
* `[256]`: Jumlah bin histogram.
* `[0, 256]`: Rentang nilai intensitas piksel (0-255).


Histogram divisualisasikan menggunakan `plt.plot()`.

## Langkah 4: Normalisasi Histogram

```python
# Menormalisasi histogram
hist_normalized = hist / hist.sum()

# Menampilkan histogram yang sudah dinormalisasi
plt.plot(hist_normalized)
plt.title("Histogram Citra Grayscale (Normalisasi)")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi Normalisasi")
plt.xlim([0, 256])
plt.grid()
plt.show()
```

* Histogram dinormalisasi dengan membagi setiap frekuensi oleh total jumlah piksel, sehingga total frekuensinya menjadi 1.
* Normalisasi membantu membandingkan histogram antar gambar dengan ukuran yang berbeda.


## Langkah 5: Gabungan Histogram Mentah dan Normalisasi

```python
# Visualisasi histogram mentah dan normalisasi berdampingan
plt.figure(figsize=(12, 5))

# Histogram mentah
plt.subplot(1, 2, 1)
plt.plot(hist)
plt.title("Histogram Mentah")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")
plt.xlim([0, 256])
plt.grid()

# Histogram normalisasi
plt.subplot(1, 2, 2)
plt.plot(hist_normalized)
plt.title("Histogram Normalisasi")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi Normalisasi")
plt.xlim([0, 256])
plt.grid()

plt.tight_layout()
plt.show()
```

* Dua subplot dibuat untuk membandingkan histogram mentah dan yang sudah dinormalisasi.
* `plt.subplot(1, 2, X)` digunakan untuk membagi area plot menjadi dua kolom.
