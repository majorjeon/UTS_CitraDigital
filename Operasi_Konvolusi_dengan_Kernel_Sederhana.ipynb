# Langkah 1: Import Library

```python
# Import library yang dibutuhkan
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

* `cv2` untuk operasi citra.
* `numpy` digunakan untuk membuat kernel rata-rata.
* `matplotlib.pyplot` untuk menampilkan gambar.


# Langkah 2: Membaca Gambar

```python
# Membaca citra grayscale
image_path = '/content/drives/MyDrive/luffy kasep.jpg'  # Ganti dengan path gambar kamu
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Periksa apakah gambar berhasil dibaca
if image is None:
    raise ValueError("Gambar tidak dapat dibaca. Pastikan path gambar benar.")

# Menampilkan gambar asli
plt.imshow(image, cmap='gray')
plt.title("Gambar Asli")
plt.axis("off")
plt.show()
```

Gambar diubah ke grayscale agar proses konvolusi lebih mudah. Input `plt.imshow()` untuk menampilkan gambar asli.

# Langkah 3: Membuat Kernel Rata-rata 3x3

```python
# Membuat kernel rata-rata 3x3
kernel = np.ones((3, 3), np.float32) / 9
print("Kernel Rata-rata 3x3:")
print(kernel)
```

Kernel rata-rata 3x3 adalah matriks 3x3 di mana setiap elemen bernilai
1/9 (1 dibagi 9 elemen). Kernel ini digunakan untuk melakukan operasi rata-rata lokal pada piksel.

# Langkah 4: Operasi Konvolusi

```python
# Operasi konvolusi menggunakan filter2D
convoluted_image = cv2.filter2D(image, -1, kernel)

# Menampilkan hasil konvolusi
plt.imshow(convoluted_image, cmap='gray')
plt.title("Hasil Konvolusi dengan Kernel Rata-rata 3x3")
plt.axis("off")
plt.show()
```

`cv2.filter2D()` menerapkan operasi konvolusi dengan parameter:
* `image`: Gambar asli.
* `-1`: Kedalaman keluaran sama dengan gambar asli.
* `kernel`: Kernel rata-rata 3x3

Hasilnya adalah gambar yang dihaluskan

# Langkah 5: Perbandingan Gambar

```python
# Membandingkan gambar asli dengan hasil konvolusi
plt.figure(figsize=(12, 5))

# Gambar asli
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Gambar Asli")
plt.axis("off")

# Gambar hasil konvolusi
plt.subplot(1, 2, 2)
plt.imshow(convoluted_image, cmap='gray')
plt.title("Hasil Konvolusi")
plt.axis("off")

plt.tight_layout()
plt.show()
```

Dua subplot dibuat untuk menampilkan gambar asli dan hasil konvolusi secara berdampingan. `plt.subplot()` digunakan untuk membagi area tampilan.

Efek Setelah Konvolusi
1. Efek yang Terjadi:
* Gambar menjadi lebih halus.
* Detail kecil seperti noise atau tepi tajam menjadi kurang terlihat.

2. Penjelasan Efek:
* Kernel rata-rata 3x3 menghitung rata-rata nilai piksel di sekitar setiap piksel, menghasilkan efek penghalusan atau blur.
* Penghalusan membantu mengurangi noise pada gambar tetapi juga dapat mengurangi ketajaman tepi.
