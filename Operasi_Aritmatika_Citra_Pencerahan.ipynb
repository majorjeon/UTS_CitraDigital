# Langkah 1: Import Library

```python
# Import library yang dibutuhkan
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Langkah 2: Membaca Citra

```python
# Membaca citra grayscale
image_path = '/content/drives/MyDrive/okarunn.jpg'  # Ganti dengan path gambar kamu
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Periksa apakah gambar berhasil dibaca
if image is None:
    raise ValueError("Gambar tidak dapat dibaca. Pastikan path gambar benar.")

# Menampilkan gambar asli
plt.imshow(image, cmap='gray')
plt.title("Gambar Asli")
plt.axis("off")
plt.show()
```

# Langkah 3: Fungsi Pencerahan Citra

```python
# Fungsi pencerahan citra
def brighten_image(image, brightness_increase):
    # Menambahkan konstanta pada setiap piksel
    brightened_image = image + brightness_increase
    # Operasi clipping untuk menjaga nilai piksel tetap antara 0-255
    brightened_image = np.clip(brightened_image, 0, 255).astype(np.uint8)
    return brightened_image

# Konstanta untuk pencerahan
brightness_increase = 50

# Menerapkan fungsi pencerahan
brightened_image = brighten_image(image, brightness_increase)

# Menampilkan hasil pencerahan
plt.imshow(brightened_image, cmap='gray')
plt.title(f"Gambar Setelah Pencerahan (+{brightness_increase})")
plt.axis("off")
plt.show()
```

Operasi pencerahan dilakukan dengan menambahkan nilai konstanta pada setiap piksel. `np.clip()` memastikan nilai piksel tetap berada dalam rentang valid [0,255] Nilai hasil akhirnya di-cast ke tipe `uint8` untuk mempertahankan format citra.

# Langkah 4: Perbandingan Gambar Asli dan Hasil Pencerahan

```python
# Membandingkan gambar asli dan hasil pencerahan
plt.figure(figsize=(12, 5))

# Gambar asli
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Gambar Asli")
plt.axis("off")

# Gambar setelah pencerahan
plt.subplot(1, 2, 2)
plt.imshow(brightened_image, cmap='gray')
plt.title(f"Gambar Setelah Pencerahan (+{brightness_increase})")
plt.axis("off")

plt.tight_layout()
plt.show()
```

Dua subplot ditampilkan untuk membandingkan gambar asli dengan gambar yang sudah diberi efek pencerahan. Gambar setelah pencerahan seharusnya terlihat lebih cerah.


# **Perubahan yang Terjadi pada Citra**
1. Efek Pencerahan:
* Nilai intensitas setiap piksel meningkat sebesar konstanta yang ditentukan, membuat gambar terlihat lebih cerah.
2. Operasi Clipping:
* Piksel dengan nilai lebih dari 255 setelah penambahan dikembalikan ke 255, mencegah overflow.
3. Pengaruh Visual:
* Area gelap menjadi lebih terang.
* Detail dalam area terang bisa hilang jika banyak piksel yang mencapai nilai maksimum (255).
